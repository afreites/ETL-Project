```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "listings-2.csv"
airbnb_df = pd.read_csv(csv_file)
airbnb_df.head()
```

```python
airbnb_data_df = airbnb_df[['zipcode', 'property_type', 'room_type','price', 'latitude', 'longitude']].copy()
```

```python
airbnb_data_df = airbnb_data_df.rename(columns={'zipcode': 'postal_code'})
airbnb_data_df.head()
```

```python
airbnb_clean_df = airbnb_data_df.loc[airbnb_data_df["postal_code"] .str.len() == 7, [
    "property_type", "room_type", "price","postal_code", 'latitude', 'longitude']]
airbnb_clean_df.head()
```

```python
airbnb_clean_df.info()
```

```python
airbnb_transformed= airbnb_clean_df.copy()
airbnb_transformed.drop_duplicates("postal_code", inplace=True)
airbnb_transformed.set_index("postal_code", inplace=True)

airbnb_transformed.head()
```

```python
airbnb_transformed.info()
```

```python
airbnb_transformed.to_excel("airbnb_list.xlsx", index=True)
```
